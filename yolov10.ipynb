```python
import shutil
import os

def remove_all_contents(directory_path):
    try:
        # List all files and directories in the specified directory
        for filename in os.listdir(directory_path):
            file_path = os.path.join(directory_path, filename)
            # Check if it is a file or directory
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)  # Remove the file or link
            elif os.path.isdir(file_path):
                shutil.rmtree(file_path)  # Remove the directory and its contents
        print(f"All contents removed from directory: {directory_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
directory_path = '/kaggle/working/'
remove_all_contents(directory_path)
```

### build dependency

```python
import subprocess
import os

def install_editable_mode(package_dir):
    try:
        # Change the current working directory to the package directory
        os.chdir(package_dir)
        
        # Run the pip install command
        result = subprocess.run(['pip', 'install', '-e', '.'], capture_output=True, text=True)
        
        # Check if the command was successful
        if result.returncode == 0:
            print(f"Successfully installed package in editable mode from {package_dir}")
        else:
            print(f"Failed to install package. Error: {result.stderr}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
package_directory = '/kaggle/working/yolov10'
install_editable_mode(package_directory)
```

```python
import subprocess

def uninstall_library(library_name):
    try:
        subprocess.check_call(['pip', 'uninstall', '-y', library_name])
        print(f'Successfully uninstalled {library_name}')
    except subprocess.CalledProcessError:
        print(f'Error uninstalling {library_name}')

# Example usage
library_name = 'wandb'  # Replace with the name of the library you want to uninstall
uninstall_library(library_name)
```

```python
import shutil
import os

def zip_folder(folder_path, zip_name):
    # Ensure the zip name does not include the '.zip' extension
    if zip_name.endswith('.zip'):
        zip_name = zip_name[:-4]
        
    # Create a zip file from the folder
    shutil.make_archive(zip_name, 'zip', folder_path)
    print(f"Folder {folder_path} has been zipped into {zip_name}.zip")

# Example usage
folder_to_zip = '/kaggle/working/yolov10/runs/detect/train'
output_zip_file = '/kaggle/working/weights'
zip_folder(folder_to_zip, output_zip_file)
```
